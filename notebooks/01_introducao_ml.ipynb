## 1. Introdução: O que é Machine Learning?

---

### 🧠 A Revolução do Aprendizado Automático

Imagine que você quer ensinar um computador a reconhecer fotos de cachorros e gatos.

**Na programação tradicional**, você teria que escrever regras como:
- _"Se tem focinho longo E orelhas pontudas E cauda enrolada → É cachorro"_
- _"Se tem bigodes E pupilas verticais E ronrona → É gato"_

Mas e as exceções? E as raças exóticas? Essa abordagem rapidamente se torna impossível.

**Com Machine Learning**, você simplesmente mostra **milhares de fotos rotuladas** ao algoritmo e ele descobre os padrões sozinho!

---

### 🔄 A Inversão de Paradigma

```
╔══════════════════════════════════════════════════════════════╗
║  PROGRAMAÇÃO TRADICIONAL                                     ║
║  Regras (escritas pelo programador) + Dados → Respostas      ║
╠══════════════════════════════════════════════════════════════╣
║  MACHINE LEARNING                                            ║
║  Dados + Respostas (exemplos) → Regras (aprendidas)          ║
╚══════════════════════════════════════════════════════════════╝
```

Esta inversão é fundamental: em vez de **programar as regras**, nós **fornecemos os exemplos** e deixamos o algoritmo descobrir as regras por conta própria.

---

### 📖 Definição Formal

> **Machine Learning** (Aprendizado de Máquina) é um subcampo da Inteligência Artificial onde algoritmos **aprendem padrões a partir de dados**, melhorando seu desempenho em tarefas específicas com a experiência, **sem serem explicitamente programados para cada regra**.
>
> — Adaptado de Tom Mitchell (1997)

---

### 🌳 ML dentro da Inteligência Artificial

```
Inteligência Artificial (IA)
└── Machine Learning (ML)
    ├── Deep Learning (Redes Neurais Profundas)
    │   ├── CNN (Visão Computacional)
    │   ├── RNN / LSTM (Séries Temporais)
    │   └── Transformers (NLP, LLMs como ChatGPT)
    ├── Aprendizado Supervisionado
    ├── Aprendizado Não Supervisionado
    └── Aprendizado por Reforço
```

## 2. Teoria: Tipos de Aprendizado e Processo de ML

---

### 2.1 O que é Machine Learning?

Machine Learning (Aprendizado de Máquina) é um subcampo da Inteligência Artificial onde algoritmos **aprendem padrões a partir de dados**, sem serem explicitamente programados.

**Programação Tradicional:** Regras + Dados → Respostas  
**Machine Learning:** Dados + Respostas → Regras (aprendidas automaticamente)

---

### 2.2 Tipos de Aprendizado

| Tipo | Descrição | Exemplos Reais |
|------|-----------|----------------|
| **Supervisionado** | Aprende com exemplos rotulados (X → y) | Previsão de crédito, diagnóstico médico |
| **Não Supervisionado** | Descobre padrões sem rótulos | Segmentação de clientes, detecção de anomalias |
| **Por Reforço** | Aprende por tentativa e erro com recompensas | Jogos, robótica autônoma |
| **Semi-supervisionado** | Combina rotulados e não rotulados | Reconhecimento de fala, NLP |

---

### 2.3 Processo Completo de ML

1. **Definição do Problema** → O que queremos prever/descobrir?
2. **Coleta de Dados** → Fontes, volume, qualidade
3. **Pré-processamento** → Limpeza, normalização, encoding
4. **Feature Engineering** → Criação e seleção de variáveis
5. **Escolha do Modelo** → Algoritmo adequado ao problema
6. **Treinamento** → Ajuste dos parâmetros do modelo
7. **Avaliação** → Métricas de performance
8. **Deploy** → Colocar o modelo em produção
9. **Monitoramento** → Acompanhar degradação do modelo

---

### 2.4 Conceitos-Chave

| Termo | Significado |
|-------|-------------|
| **Features (X)** | Variáveis de entrada / atributos dos dados |
| **Target (y)** | Variável que queremos prever (saída) |
| **Modelo** | Função matemática aprendida dos dados |
| **Treinamento** | Processo de ajuste dos parâmetros do modelo |
| **Overfitting** | Modelo memoriza os dados, não generaliza |
| **Underfitting** | Modelo muito simples, não captura os padrões |
| **Validação** | Conjunto de dados para ajuste de hiperparâmetros |
| **Teste** | Conjunto nunca visto para avaliação final |

````python
# ============================================================
# CÉLULA 4: Visualização dos 4 Paradigmas de Machine Learning
# ============================================================
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
import os

np.random.seed(42)
fig = plt.figure(figsize=(14, 10))
gs = GridSpec(2, 2, figure=fig, hspace=0.4, wspace=0.35)

# --- Supervisionado: Regressão ---
ax1 = fig.add_subplot(gs[0, 0])
x = np.linspace(0, 10, 60)
y = 2.5 * x + np.random.randn(60) * 3
ax1.scatter(x, y, alpha=0.6, color='steelblue', s=40, label='Dados observados')
ax1.plot(x, 2.5*x, 'r-', lw=2.5, label='Modelo aprendido')
ax1.set_title('Aprendizado Supervisionado\n(Regressão — Preço de Imóveis)', fontsize=10, fontweight='bold')
ax1.set_xlabel('Área (m²)', fontsize=9); ax1.set_ylabel('Preço (R$ mil)', fontsize=9)
ax1.legend(fontsize=8); ax1.grid(True, alpha=0.3)

# --- Supervisionado: Classificação ---
ax2 = fig.add_subplot(gs[0, 1])
X_pos = np.random.randn(50, 2) + [2, 2]
X_neg = np.random.randn(50, 2) + [-2, -2]
ax2.scatter(X_pos[:,0], X_pos[:,1], c='#e74c3c', s=50, alpha=0.7, label='Inadimplente')
ax2.scatter(X_neg[:,0], X_neg[:,1], c='#2ecc71', s=50, alpha=0.7, label='Adimplente')
xx = np.linspace(-5, 5, 100)
ax2.plot(xx, -xx, 'navy', lw=2.5, linestyle='--', label='Fronteira de decisão')
ax2.set_title('Aprendizado Supervisionado\n(Classificação — Score de Crédito)', fontsize=10, fontweight='bold')
ax2.legend(fontsize=8); ax2.grid(True, alpha=0.3)
ax2.set_xlim(-5, 6); ax2.set_ylim(-5, 6)

# --- Não Supervisionado: Clustering ---
ax3 = fig.add_subplot(gs[1, 0])
centers = [[1,1],[5,5],[-2,5]]
palette = ['#e74c3c', '#3498db', '#f39c12']
for i, (c, col) in enumerate(zip(centers, palette)):
    pts = np.random.randn(40, 2) * 0.8 + c
    ax3.scatter(pts[:,0], pts[:,1], c=col, s=50, alpha=0.7, label=f'Segmento {i+1}')
    ax3.scatter(*c, c='black', s=200, marker='*', zorder=5)
ax3.set_title('Aprendizado Não Supervisionado\n(Clustering — Segmentação de Clientes)', fontsize=10, fontweight='bold')
ax3.legend(fontsize=8); ax3.grid(True, alpha=0.3)

# --- Aprendizado por Reforço ---
ax4 = fig.add_subplot(gs[1, 1])
episodios = np.arange(1, 101)
recompensa = -50 * np.exp(-episodios/30) + 45 + np.random.randn(100) * 3
ax4.plot(episodios, recompensa, color='#9b59b6', lw=2, label='Recompensa acumulada')
ax4.axhline(y=40, color='red', linestyle='--', alpha=0.7, label='Meta atingida')
ax4.fill_between(episodios, recompensa, alpha=0.15, color='#9b59b6')
ax4.set_title('Aprendizado por Reforço\n(Agente aprendendo a jogar)', fontsize=10, fontweight='bold')
ax4.set_xlabel('Episódio de treinamento', fontsize=9)
ax4.set_ylabel('Recompensa acumulada', fontsize=9)
ax4.legend(fontsize=8); ax4.grid(True, alpha=0.3)

plt.suptitle('Os 4 Paradigmas de Machine Learning', fontsize=14, fontweight='bold', y=1.01)

# Salvar imagem
os.makedirs('../imagens', exist_ok=True)
plt.savefig('../imagens/01_tipos_de_ml.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Gráfico salvo em ../imagens/01_tipos_de_ml.png')
````

````python
# ============================================================
# CÉLULA 6: Demo — Programação Tradicional vs Machine Learning
# Classificação de e-mails: SPAM vs NÃO-SPAM
# ============================================================
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print('=' * 60)
print('ABORDAGEM 1: Programação Tradicional (Regras Hardcoded)')
print('=' * 60)

def classificar_spam_tradicional(email):
    """Regras escritas manualmente por um programador."""
    palavras_spam = ['promoção', 'grátis', 'ganhe', 'clique aqui', 'oferta', 'urgente']
    email_lower = email.lower()
    pontuacao = sum(1 for p in palavras_spam if p in email_lower)
    return 'SPAM' if pontuacao >= 2 else 'NÃO-SPAM'

emails_teste = [
    'Promoção grátis! Ganhe R$1000 clicando aqui agora!',
    'Reunião amanhã às 14h para discutir o projeto.',
    'Oferta exclusiva! Promoção por tempo limitado!',
    'Relatório mensal em anexo conforme solicitado.',
]

print('\nClassificação por Regras:')
for email in emails_teste:
    resultado = classificar_spam_tradicional(email)
    icone = '🚨' if resultado == 'SPAM' else '✅'
    print(f'  {icone} [{resultado}] {email[:50]}...')

print('\n⚠️  Problema: E se aparecer "desconto imperdível"? Ou spam em inglês?')
print('   As regras precisariam ser atualizadas manualmente SEMPRE.\n')

print('=' * 60)
print('ABORDAGEM 2: Machine Learning (Aprende dos Dados)')
print('=' * 60)

# Dataset de treinamento
emails_treino = [
    ('Promoção incrível! Ganhe dinheiro fácil!', 1),
    ('Clique aqui para oferta exclusiva grátis!', 1),
    ('Você ganhou um prêmio! Resgate urgente!', 1),
    ('Reunião amanhã às 10h na sala de conferências.', 0),
    ('Segue o relatório trimestral conforme solicitado.', 0),
    ('Parabéns pelo seu aniversário! Abraços.', 0),
    ('Oferta especial grátis só hoje! Não perca!', 1),
    ('Atualização do sistema enviada para aprovação.', 0),
    ('Ganhe dinheiro rápido sem sair de casa!', 1),
    ('Sua fatura do cartão está disponível.', 0),
]

textos = [e[0] for e in emails_treino]
rotulos = [e[1] for e in emails_treino]

# Treinar o modelo
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(textos)
modelo_ml = MultinomialNB()
modelo_ml.fit(X, rotulos)

print(f'\n📚 Modelo treinado com {len(emails_treino)} exemplos')
print('\nClassificação por ML (treinado em 10 exemplos):')
for email in emails_teste:
    X_teste = vectorizer.transform([email])
    pred = modelo_ml.predict(X_teste)[0]
    proba = modelo_ml.predict_proba(X_teste)[0]
    resultado = 'SPAM' if pred == 1 else 'NÃO-SPAM'
    confianca = max(proba) * 100
    icone = '🚨' if pred == 1 else '✅'
    print(f'  {icone} [{resultado} ({confianca:.0f}% confiança)] {email[:45]}...')

print('\n💡 Diferença chave: ML aprendeu os padrões automaticamente dos dados!')
print('   Basta adicionar mais exemplos para melhorar o modelo.')
print('\n📊 Comparação:')
comparacao = pd.DataFrame({
    'Aspecto': ['Desenvolvimento', 'Manutenção', 'Adaptação', 'Escalabilidade', 'Novas linguagens/padrões'],
    'Programação Tradicional': ['Programador escreve regras', 'Manual (sempre)', 'Difícil', 'Limitada', 'Requer reescrita'],
    'Machine Learning': ['Algoritmo aprende dos dados', 'Re-treinar com novos dados', 'Automática', 'Alta', 'Basta adicionar exemplos']
})
print(comparacao.to_string(index=False))
````

````python
# ============================================================
# CÉLULA 8: Demo Completo — Como o Modelo APRENDE com os Dados
# Regressão Linear: Previsão de Preço de Apartamentos em SP
# ============================================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os

np.random.seed(42)

# Simula dados de área x preço de apartamentos em SP
area = np.random.uniform(30, 200, 100).reshape(-1, 1)
preco = 8000 * area.flatten() + np.random.randn(100) * 50000 + 200000

print('🏙️  Dataset: Preço de Apartamentos em São Paulo')
print(f'   Total de imóveis: {len(area)}')
print(f'   Área média: {area.mean():.0f} m²')
print(f'   Preço médio: R$ {preco.mean():,.0f}')
print(f'   Coeficiente REAL: R$ 8.000/m²\n')

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

tamanhos_treino = [5, 20, 100]
titulos = ['Poucos dados (5 imóveis)', 'Dados medianos (20 imóveis)', 'Dados suficientes (100 imóveis)']
cores = ['#e74c3c', '#f39c12', '#2ecc71']

for ax, n, titulo, cor in zip(axes, tamanhos_treino, titulos, cores):
    area_n = area[:n]
    preco_n = preco[:n]
    
    modelo = LinearRegression()
    modelo.fit(area_n, preco_n)
    
    x_linha = np.linspace(25, 210, 200).reshape(-1, 1)
    y_pred = modelo.predict(x_linha)
    
    ax.scatter(area_n, preco_n/1000, alpha=0.7, color=cor, s=60, label='Dados de treino', zorder=5)
    ax.plot(x_linha, y_pred/1000, 'navy', lw=2.5, label=f'Modelo (coef={modelo.coef_[0]:.0f} R$/m²)')
    ax.axline((0, modelo.intercept_/1000), slope=8, color='gray', lw=1.5,
              linestyle='--', alpha=0.6, label='Valor real (R$8.000/m²)')
    ax.set_title(titulo, fontweight='bold', fontsize=11)
    ax.set_xlabel('Área (m²)', fontsize=9)
    ax.set_ylabel('Preço (R$ mil)', fontsize=9)
    ax.legend(fontsize=7.5)
    ax.grid(True, alpha=0.3)
    
    r2 = modelo.score(area_n, preco_n)
    erro_coef = abs(modelo.coef_[0] - 8000)
    ax.text(0.05, 0.95, f'R² = {r2:.3f}\nErro no coef: R${erro_coef:.0f}/m²',
            transform=ax.transAxes, fontsize=8.5, va='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.suptitle('Como o Modelo "Aprende" com Mais Dados\n(Previsão de Preço de Apartamentos em SP)',
             fontsize=13, fontweight='bold')
plt.tight_layout()

os.makedirs('../imagens', exist_ok=True)
plt.savefig('../imagens/01_aprendizado_modelo.png', dpi=150, bbox_inches='tight')
plt.show()

m5   = LinearRegression().fit(area[:5], preco[:5]).coef_[0]
m20  = LinearRegression().fit(area[:20], preco[:20]).coef_[0]
m100 = LinearRegression().fit(area, preco).coef_[0]

print(f'\n📊 Coeficiente verdadeiro: R$ 8.000/m²')
print(f'   Com  5 dados:   R$ {m5:.0f}/m²  → Erro de {abs(m5-8000):.0f} R$/m²  ({abs(m5-8000)/8000*100:.1f}%)')
print(f'   Com 20 dados:   R$ {m20:.0f}/m²  → Erro de {abs(m20-8000):.0f} R$/m²   ({abs(m20-8000)/8000*100:.1f}%)')
print(f'   Com 100 dados:  R$ {m100:.0f}/m²  → Erro de {abs(m100-8000):.0f} R$/m²    ({abs(m100-8000)/8000*100:.1f}%)')
print('\n💡 Com mais dados, o modelo converge para o coeficiente real!')
print('   Esta é a essência do Machine Learning: aprender padrões dos dados.')
````

### 🎯 O que aprendemos neste módulo

1. **Machine Learning é uma inversão de paradigma**: em vez de programar regras manualmente, fornecemos exemplos e o algoritmo descobre os padrões sozinho.

2. **Existem 4 paradigmas principais** de aprendizado:
   - **Supervisionado**: aprende com exemplos rotulados (classificação e regressão)
   - **Não Supervisionado**: descobre estruturas ocultas nos dados
   - **Por Reforço**: aprende por tentativa, erro e recompensa
   - **Semi-supervisionado**: combina os dois primeiros

3. **O processo de ML é iterativo** e envolve 9 etapas: da definição do problema até o monitoramento em produção.

4. **Mais dados geralmente significa melhor aprendizado**, mas qualidade supera quantidade.

5. **ML já está no seu dia a dia**: Netflix, iFood, Nubank, Waze e centenas de outros serviços usam ML continuamente.

---

### 📊 Tabela Comparativa dos Tipos de ML

| Característica | Supervisionado | Não Supervisionado | Por Reforço |
|----------------|---------------|-------------------|-------------|
| **Dados necessários** | Rotulados (X + y) | Não rotulados (apenas X) | Ambiente + função de recompensa |
| **Tipo de saída** | Classe ou valor numérico | Grupos / estruturas | Ação a tomar |
| **Feedback** | Imediato (erro no treino) | Nenhum direto | Atrasado (recompensa) |
| **Dificuldade** | Média | Alta | Muito alta |
| **Custo de dados** | Alto (rotulação manual) | Baixo (sem rótulos) | Médio (simulação) |
| **Aplicações típicas** | Fraude, diagnóstico, preço | Segmentação, anomalias | Jogos, robótica, trading |

### 🔑 Terminologia Essencial

```
Dataset  →  Conjunto de dados usado para treinar/avaliar o modelo
Features →  Variáveis de entrada (colunas X)
Target   →  Variável que queremos prever (coluna y)
Modelo   →  Função matemática aprendida dos dados
Treino   →  Dados usados para ajustar o modelo (~70-80%)
Teste    →  Dados nunca vistos para avaliar o modelo (~20-30%)
R²       →  Coeficiente de determinação (0 = ruim, 1 = perfeito)
```
